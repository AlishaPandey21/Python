# List Comprehension

```python
z=[]
l=[1,2,3,4]
for i in l:
    z.append(i**2)
print(z)

z1=[i**2 for i in l]
print(z1)
z2=[i**2 for i in l if i%2==0]
print(z2)
z3=[i**2 if i%2==0 else i**3 for i in l]
print(z3)
```

# Min and Max in List

```python
l=[]
el=[]
ol=[]
for i in range (10):
    x=eval(input("Enter Digit"))
    l.append(x)
for i in l:
    if i%2==0:
        el.append(i)
    else:
        ol.append(i)
l.sort()
print("Max Number",l[-1])
print("Min Number",l[0])
print("second Max Number",l[-2])
print("Min Number",l[1])
print("Max Even Number is",el[-1])
print("Max Odd Number",ol[-1])
print("Min Even Number is",el[0])
print("Min Odd Number",ol[0])
```

# Longest SubSequence

```python

x=["pyt","pytho","python","pyth"]
x.sort(reverse=True)
ans=""

for i in range (len (x[-1])):
    if x[0][i]==x[-1][i]:
        ans=ans+x[-1][i]
        continue
    else:
        print("String Not Found")
        break;
print("Longest Subsequence String is: ",ans)
```
