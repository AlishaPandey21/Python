# File Operations


```python
# f=open(filename,mode)
f=open("1.txt") #BY DEFAULT IN READ MODE
x=f.read()
f.close()
print(x)
```

```python
f=open("1.txt","r")#Gives Permission to read only
x1=f.read()
f.close()#closing is necessary otherwise it will not let to perform other operation eficiently
x1
```

```python
f=open("1.txt","r+")
#Gives Permission to read as well as write
#Also the pointer point at beginning of the file only 
f.write("AAA")
x=f.read()#Pointer is at index 3
f.close()
print(x)
```

```python
f=open("1.txt","r+")
f.write("ABC")#will Overwrite
f.write("DEF")
f.seek(0)#to make pointer reach at begin of the file otherwise it will read from the last position of pointer
f.read()
```
