```python
class circle:
    def __init__(self,radius):
        self.radius=radius
    def aoc(self):
        print("area of circle is",3.14*self.radius**2)
    def poc(self):
        print("perimeter of circle is",3.14*self.radius*2)
d=circle(1)
d.aoc()
d.poc()
```
