# Functions

(Map-Filter-Reduce)

```python
l=[1,2,3,4]
z=list(map(lambda i:i+1,l))
print(z)
y=tuple(filter(lambda i:i%2==0,l))
print(y)
```

```python
from functools import *
x=reduce(lambda x, y:x*y,l)
print(x)
```

```python
L1=[2024,2025,2026]
f1=list(filter(lambda i:i==2024,L1))
print(f1)
```

# maketrans and translate

```python
s="Hello Have a Nice day"
s.replace("0","@")
ans=str.maketrans("HNd","@#A","n")
print(ans)
sol=s.translate(ans)
print(sol)
```

```python
s="he@# INDIA is =.921"
ans=str.maketrans("I","i","@#=.")
print(ans)
sol=s.translate(ans)
print(sol)
```

```python
import string
z=string.punctuation
print(z)
s="he@# INDIA is =.921"
ans=str.maketrans("I","i",z)

sol=s.translate(ans)
print(sol)
```
