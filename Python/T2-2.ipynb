```python
x=input("Enter String")
encrypt=int(input("Enter no. of encryption"))
x=x.upper()
Final=""
ans="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in x:
    if i in ans:
        op=(ans.find(i)+encrypt)%26
        Final=Final+ans[op]
    else:
        Final=Final+i
        
print(Final)    
```
