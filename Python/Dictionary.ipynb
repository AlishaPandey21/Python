#  players name with max and min by runs

```python
d={}
for i in range(2):
    name=input("Enter Name of the Player")
    score=int(input("Enter Score of the player"))
    d.update({name:score})
x=max(d.values())
min1=min(d.values())



for i,j in d.items():
    if j==x:
        print(f"Max Score is {j} by {i}")
    if j==min1:
        print(f"Min Score is {j} by {i}")
      
z=sum(d.values())/2
print("Avg",z)





    
```

# number to spelling

```python
d={1:"One",2:"Two",3:"Three",4:"Four",5:"Five",6:"Six",7:"Seven",8:"Eight",9:"Nine",10:"Ten",11:"Eleven",
   12:"Twelve",13:"Thirteen",14:"Fourteen",15:"Fifteen",16:"Sixteen",17:"Seventeen",18:"Eighteen",19:"Nineteen",20:"Twenty",
   30:"Thirty",40:"Fourty",50:"Fifty",60:"Sixty"}
n=eval(input("Enter Number"))
def num1(n):
    if n<=20:
        
         return d[n]
    if  n<100:
        if n%10==0:
            return d[n]
        else:
            return d[(n//10)*10]+d[n%10]
    if n<1000:
        if n%100==0:
            return d[n//100]+"Hundred"
        else:
            return d[n//100]+"Hundred"+num1(n%100)
        
        
        
z=num1(n)
print(z)
    
    
    

```
